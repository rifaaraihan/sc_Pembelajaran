```python
def rating_group(rating):
    if rating >=7.5:
        return 'Good'
    elif rating >= 6.0:
        return 'Medium'
    else:
        return 'Baad'
```

```python
data['Rating_category'] = data['Rating'].apply(rating_group)
```
